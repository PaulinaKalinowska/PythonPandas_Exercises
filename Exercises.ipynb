```python
#indexing
import string
litery = list(string.ascii_uppercase)
dane3 = pd.Series(losowe, index=litery)
print(dane3.head())
```

```python
print(dane3['E'])
print(dane3[['P', 'Y', 'T']])
print(dane3[0:7])
```

```python
#maping
def cube(x):
    return str(x)
print(dane3.map(cube).head())
print(dane3.head())
```
